```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('day.csv')
df.set_index('instant', inplace=True)
print(df.head())
```

```python
df['dteday'] = pd.to_datetime(df['dteday'])
print(df.dtypes)
print(df.temp.describe())
print(df.cnt.describe())
```

```python
plt.scatter(x=df['cnt'], y=df['atemp'], s=1)
plt.xlabel('Count of bike rides')
plt.ylabel('Temperature')
plt.show()
```

```python
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.plot(df['cnt'], color=color)
ax1.set_xlabel('day of program')
ax1.set_ylabel('counts of bike rentals', color=color)

color = 'tab:red'
ax2 = ax1.twinx()
ax2.plot(df['atemp'], color='red')
ax2.set_ylabel('normalized feel temperature', color=color)
plt.show()
```

```python
plt.boxplot(df['cnt'])
plt.show()
```

The boxplot shows no outliers.  Usually between 3000-6000 bike rides.

```python
df_month = df.groupby(by=[df.yr, df.mnth]).agg({'cnt': 'sum', 'temp': 'mean'})
```

```python
print(df_month)
```

```python
plt.hist(df_month['cnt'])
plt.show()
```
